On observe 94 codes de cancer pour 153 pays différents entre 1950 et 2015. 

```python
# Evolution des observations par année
Nb_observation_by_year = mortality.groupby(['year'])['Number of deaths'].count()
Nb_observation_by_year.transpose().plot(kind='line',figsize=(25, 10))
plt.title('Nombre d observations par année', bbox={'facecolor':'0.8', 'pad':0})
```

```python
# Evolution du nombre de mort par année
Nb_mortality_by_year = mortality.groupby(['year'])['Number of deaths'].sum()
Nb_mortality_by_year.transpose().plot(kind='line',figsize=(25, 10))
plt.title('Nombre de mort par année', bbox={'facecolor':'0.8', 'pad':0})
```

```python
# Evolution des observations par pays
Nb_observation_by_location = mortality.groupby(['location'])['Number of deaths'].count()
Nb_observation_by_location.transpose().plot(kind='bar',figsize=(25, 10))
plt.title('Nombre d observations par pays', bbox={'facecolor':'0.8', 'pad':0})
```

```python
# Evolution du nombre de morts par pays
Nb_mortality_by_location = mortality.groupby(['location'])['Number of deaths'].sum()
Nb_mortality_by_location.transpose().plot(kind='bar',figsize=(25, 10))
plt.title('Nombre de mort par pays', bbox={'facecolor':'0.8', 'pad':0})
```

# Fusion des deux bases de données : "mortality" et "world bank"

On souhaite fusionner les deux bases de données par année et pays. Ensuire, on devra restreindre la base fusionnée entre 1970 et 2015 (l'intersection des deux bases). De plus, on devra égalemment supprimer les pays qui ne sont pas présents dans la base world bank car on aura pas de données socio - économiques. Donc on utilise how='inner'.

# Attention ne marche pas quand on met la base worldbank en entier (trop de colonne), mais cela marche quand on réduit le nombre de colonne

```python
# l'opération suivante correspond à un JOIN (en SQL)
#Base_fusion = t2012.merge(t2014, left_on="Catégorie d'emploi",  right_on ="Catégorie d'emploi",how='outer')
Base_fusion = pd.merge(mortality, worldbank, how='inner',left_on=['location', 'year'],right_on=['area', 'year'])
#Base_fusion = pd.merge(mortality, worldbank, how='inner',on=['area', 'year'])
#Base_fusion = pd.merge(mortality, worldbank, on='year')
```
